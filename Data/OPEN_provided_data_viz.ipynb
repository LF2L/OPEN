# Visualisation des données fournies avec OPEN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Production photovoltaique en Juin

```python
PV_data_path = os.path.join("Building/", "PVpu_1min_2013JUN.csv")    
PVpu_raw_smr = pd.read_csv(PV_data_path, index_col=0, parse_dates=True).values
PVtotal_smr = np.sum(PVpu_raw_smr,1)
print(PVtotal_smr.shape)

plt.plot(PVtotal_smr)
plt.show()
```
